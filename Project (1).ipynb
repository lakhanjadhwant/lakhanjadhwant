Data Cleaning:- Category , Rating , Reviews , Size , Installs, Price, Last Updated

```python
dfgply['Category'].unique()
```

```python
dfgply[dfgply['Category'] == '1.9']
```

```python
dfgply.loc[10472] = dfgply.loc[10472].shift()
dfgply['App'].loc[10472] = dfgply['Category'].loc[10472]
dfgply['Category'].loc[10472] = np.nan
dfgply.loc[10472]
```

```python
dfgply['Rating'].unique()
```

```python
dfgply['Rating'] = pd.to_numeric(dfgply['Rating'], errors='coerce')
dfgply['Rating'].dtype
```

```python
dfgply['Reviews'].unique()
```

```python
dfgply[dfgply['Reviews'] =='3.0M']
```

```python
dfgply['Reviews'] = dfgply.Reviews.replace("0.0",0)
dfgply['Reviews'] = dfgply.Reviews.replace("3.0M",3000000.0)
dfgply['Reviews'] = dfgply['Reviews'].astype(float)
dfgply['Reviews'].dtype
```

```python
dfgply['Size'].unique()
```

```python
dfgply['Size'] = dfgply.Size.replace("Varies with device",np.nan)
dfgply['Size'] = dfgply.Size.str.replace("M","000")          # All size values became the kilobyte type.
dfgply['Size'] = dfgply.Size.str.replace("k","")
dfgply['Size'] = dfgply.Size.replace("1,000+",1000)
dfgply['Size'] =dfgply['Size'].astype(float)
dfgply['Size'].dtype
```

```python
dfgply['Installs'].unique()
```

```python
dfgply['Installs'] = dfgply.Installs.str.replace(",","")
dfgply['Installs'] = dfgply.Installs.str.replace("+","")
dfgply['Installs'] = dfgply.Installs.replace("Free",np.nan)
dfgply['Installs'] = dfgply['Installs'].astype(float)
dfgply['Installs'].dtype
```

```python
dfgply['Price'].unique()
```

```python
dfgply['Price'] = dfgply.Price.replace("Everyone",np.nan)
dfgply['Price'] = dfgply.Price.str.replace("$","").astype(float)
dfgply['Price'].dtype
```

```python
dfgply['Last Updated'].unique()
```

```python
dfgply['Last Updated'] = pd.to_datetime(dfgply['Last Updated'])
dfgply['Last Updated']
```

Exploratory Data Analysis:-

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,5))
sns.countplot(x='Category',data=dfgply)
plt.title('Number of Categories in App')
plt.xticks(rotation=90)
plt.ylabel('Numbers of App')
plt.show()
```

Ratings:-

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Rating', data=dfgply)
plt.title('Rating Distribution')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()
```

Installs :-

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Installs', data=dfgply)
plt.title('Installations')
plt.xticks(rotation=90)
plt.ylabel('Number of Apps')
plt.show()
```

Category and count:-

```python
plt.figure(figsize=(12,12))
most_cat = dfgply.Category.value_counts()
sns.barplot(x=most_cat, y=most_cat.index, data=dfgply)
plt.show()
```

WordCloud:-

```python
plt.subplots(figsize=(8,8))
wordcloud = WordCloud(
                          background_color='black',
                          width=512,
                          height=384
                         ).generate(" ".join(dfgply))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('graph.png')

plt.show()
```

```python
#correlation map
f,ax = plt.subplots(figsize=(8,8))
sns.heatmap(dfgply.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```
